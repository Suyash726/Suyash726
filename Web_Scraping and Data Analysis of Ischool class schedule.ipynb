```python
import pandas as pd
#To turn off warning messages
import warnings
warnings.filterwarnings('ignore')
```

```python
#Steps we need to perform!
# Getting the data from the site.
#Reading all 5 pages of the data
#Figure out which classes are grad anbd which are undergrad
#Display the undergrad 8am and undergrad Friday classes
```

```python
def get_ischool_classes():
    classes = pd.DataFrame() 

    for i in range(1,6):
        link = 'https://ischool.syr.edu/classes/?page=' + str(i)
        data = pd.read_html(link)    
        classes = classes.append(data[0], ignore_index=True)
    
    classes.columns = ['Course','Section','ClassNo','Credits','Title','Instructor','Time','Days','Room']

    return classes
```

```python
classes = get_ischool_classes()
classes.sample(5)
```

```python
#So now we have our data! Now proceeding to step 2 and figuring out which courses are undergrad! 
#As a matter of fact the courses with last three numbers >= 500 are Grad courses and those < 500 are Ugrad  
```

```python
classes['Number'] = classes['Course'].str[3:] 
classes['Type'] = ''
classes['Type'][classes['Number'] >= '500'] = 'Grad' 
classes['Type'][classes['Number'] < '500'] = 'UGrad'
```

```python
#Observe that we have a Type column now
classes.tail()
```
